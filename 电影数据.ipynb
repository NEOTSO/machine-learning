## 统计电影种类

```python
data.head()
```

```python
data['Genre']
```

```python
temp_list = [j.split(',') for j in data['Genre']]
```

```python
temp_list
```

```python
all_genre = [i for j in temp_list for i in j]
```

```python
all_genre
```

```python
genre_list = np.unique(all_genre)
genre_list
```

```python
data.shape
```

```python
genre_list.shape
```

```python
zeros = np.zeros([data.shape[0], genre_list.shape[0]])
zeros
```

```python
zeros.shape
```

```python
temp_df = pd.DataFrame(zeros, columns=genre_list)
```

```python
temp_df
```

```python
data.shape[0]
```

```python
t = pd.DataFrame(temp_df)
```

```python
t
```

```python
for i in range(data.shape[0]):
    t.iloc[[i], t.columns.get_indexer(temp_list[i])] = 1
```

```python
t
```

```python
genre = t.sum().sort_values(ascending=False)
```

```python
genre
```

```python
genre.plot(kind="bar", colormap="cool", figsize=(20, 8), fontsize=16)
```
